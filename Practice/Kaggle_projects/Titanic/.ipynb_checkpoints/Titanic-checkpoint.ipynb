В датасетах присутствуют следующие __параметры__:
* __Survived__ - цель предсказаний; 0 - погиб пассажир, 1 - пассажир выжил;
* __Pclass__ - класс пассажира на борту;
* __Name__ - имя пассажира и его социальный статус;
* __Sex__ - пол пассажира: f-женский, m-мужской;
* __Age__ - возраст пассажира;
* __SibSp__ - количесвто братьев, сестер или супругов на борту;
* __Parch__ - количество детей или родителей на борту;
* __Ticket__ - номер билета;
* __Fare__ - цена билета в британских фунтах стерлингах;
* __Cabin__ - номер каюты кассажира;
* __Embarked__ - порт, в котором пассажир поднялся на борт Титаника: C=Cherbourg, Q=Queenstown, S=Southampton.
